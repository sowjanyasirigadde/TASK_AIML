5.Operators

1. Arithmetic Operators

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a // b)
print(a **b)
```

2. Comparison Operators

```python
a = 10
b = 5

print(a == b)
print(a != b)
print(a > b)
print(a < b)
```

3. Logical Operators

```python
age = 20

print(age > 18 and age < 30)
print(age > 18 or age > 50)
print(not(age > 18))
```

4. Assignment Operators

```python
x = 10

x += 5
x -= 2
x *= 3

print(x)
```
