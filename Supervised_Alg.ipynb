SVM – Mobile Quality Check

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
data_svm = {
    'Battery_Health': [98,95,90,88,85,99,80,93,78,96],
    'Screen_Defects': [0,0,1,2,2,0,3,1,4,0],
    'Quality': [1,1,1,0,0,1,0,1,0,1]
}
```

```python
df = pd.DataFrame(data_svm)

print(df)
```

```python
X = df[['Battery_Health', 'Screen_Defects']]
y = df['Quality']
model = SVC()
model.fit(X ,y)
prediction = model.predict([[92,1]])
print("\nprediction:",prediction)
if prediction[0] == 1:
    print("Mobile Quality: Good")
else:
    print("Mobile Quality: Defective")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(x='Quality', data=df, palette='viridis', hue='Quality', legend=False)
plt.title('Mobile Quality Check (SVM) - Quality Distribution')
plt.xlabel('Quality (0=Defective, 1=Good)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Defective', 'Good']) # Label the x-axis ticks
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Polynomial Regression – Battery Charging Prediction

### Visualization for Polynomial Regression – Battery Charging Prediction
