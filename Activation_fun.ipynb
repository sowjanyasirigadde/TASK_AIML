Sigmoid


```python
import numpy as np

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Sample dataset
hours_studied = [1, 2, 4, 6, 8]

print("Student Pass Prediction")
print("-" * 40)

for hours in hours_studied:
    # Simple model
    z = 1.2 * hours - 5

    # Probability using sigmoid
    probability = sigmoid(z)

    # Decision
    if probability >= 0.5:
        result = "Pass"
    else:
        result = "Fail"

    print(f"Hours Studied: {hours}")
    print(f"Score (z): {z:.2f}")
    print(f"Pass Probability: {probability:.2f}")
    print(f"Prediction: {result}")
    print("-" * 40)
```

ReLU (Rectified Linear Unit)

```python
import numpy as np

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Market scores from the model
market_scores = [-5, -2, 0, 3, 6, 9]

print("House Price Increase Prediction")
print("-" * 40)

for score in market_scores:
    increase = relu(score)

    if increase == 0:
        result = "No Price Increase"
    else:
        result = f"House Price Increased by {increase}%"

    print(f"Market Score: {score}")
    print(f"ReLU Output: {increase}")
    print(f"Prediction: {result}")
    print("-" * 40)
```

Tanh (Hyperbolic Tangent)

```python
import numpy as np

# Tanh function
def tanh(x):
    return np.tanh(x)

# Employee performance scores
scores = [-4, -2, 0, 2, 4]

print("Employee Performance Evaluation")
print("-" * 45)

for score in scores:
    performance = tanh(score)

    if performance > 0:
        status = "Good Performance"
    elif performance < 0:
        status = "Needs Improvement"
    else:
        status = "Average"

    print(f"Performance Score: {score}")
    print(f"Tanh Output: {performance:.4f}")
    print(f"Evaluation: {status}")
    print("-" * 45)
```

Leaky ReLU

```python
import numpy as np

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Product demand scores
demand_scores = [-4, -2, 0, 3, 6]

print("Product Demand Prediction")
print("-" * 40)

for score in demand_scores:
    demand = leaky_relu(score)

    if demand > 0:
        status = "High Demand"
    elif demand == 0:
        status = "Average Demand"
    else:
        status = "Low Demand"

    print(f"Demand Score: {score}")
    print(f"Leaky ReLU Output: {demand:.2f}")
    print(f"Prediction: {status}")
    print("-" * 40)
```

Softmax


```python
import numpy as np

# Softmax function
def softmax(x):
    exp_values = np.exp(x)
    return exp_values / np.sum(exp_values)

# Model scores (logits)
scores = np.array([2.0, 1.0, 0.1])

fruits = ["Apple", "Banana", "Orange"]

probabilities = softmax(scores)

print("Fruit Classification")
print("-" * 40)

for fruit, prob in zip(fruits, probabilities):
    print(f"{fruit}: {prob:.2f}")

prediction = fruits[np.argmax(probabilities)]

print("-" * 40)
print("Predicted Fruit:", prediction)
```
