### Data Cleaning and Preprocessing

```python
# Reload the data, assuming the first row (index 0) contains the correct headers
df = pd.read_excel('Presentation score_TASK.xlsx', header=0)

# The first data row (index 0 of the DataFrame) appears to be a summary/example row.
# We will drop this row to get to the actual participant data.
df = df.iloc[1:].copy()

# Rename the 'Unnamed: 2' column to 'Group No' and ensure other relevant columns are named correctly.
# Based on df.head() from initial load, the columns are already correct, but let's confirm explicit rename for robustness.
df.columns = ['UID', 'Participant Name', 'Group No', 'Project Name', 'Technical knowledge', 'Communication', 'Attire', 'Marks']

# Remove any extra spaces from column names that might have been introduced
df.columns = df.columns.str.strip()

# Display the cleaned DataFrame head and info
print('Cleaned DataFrame Head:')
display(df.head())
print('\nCleaned Column Information:')
df.info()
```

```python
# Convert 'Technical knowledge' to numeric, coercing errors to NaN
df['Technical knowledge'] = pd.to_numeric(df['Technical knowledge'], errors='coerce')

# Check for missing values before dropping
print('Missing Values Before Dropping:')
display(df.isnull().sum()[df.isnull().sum() > 0])

# Remove rows with any missing values from the relevant columns
initial_rows = df.shape[0]
df_cleaned = df.dropna(subset=['Technical knowledge', 'Communication', 'Attire']).copy()
rows_after_drop = df_cleaned.shape[0]

print(f"\nRemoved {initial_rows - rows_after_drop} rows with missing values in 'Technical knowledge', 'Communication', or 'Attire'.")
print(f"DataFrame shape after dropping NaNs: {df_cleaned.shape}")

# Display missing values after dropping
print('\nMissing Values After Dropping:')
display(df_cleaned.isnull().sum()[df_cleaned.isnull().sum() > 0])
```

### Create 'Total Marks' and Identify Top Students

```python
# Create 'Total Marks' column
df_cleaned['Total Marks'] = df_cleaned['Technical knowledge'] + df_cleaned['Communication'] + df_cleaned['Attire']

# 1. Top 5 students based on Total Marks
top_total_marks = df_cleaned.sort_values(by='Total Marks', ascending=False).head(5)
print('Top 5 students based on Total Marks:')
display(top_total_marks[['Participant Name', 'Total Marks']])

# 2. Top 5 students in Technical knowledge
top_technical = df_cleaned.sort_values(by='Technical knowledge', ascending=False).head(5)
print('\nTop 5 students in Technical knowledge:')
display(top_technical[['Participant Name', 'Technical knowledge']])

# 3. Top 5 students in Communication
top_communication = df_cleaned.sort_values(by='Communication', ascending=False).head(5)
print('\nTop 5 students in Communication:')
display(top_communication[['Participant Name', 'Communication']])

# 4. Top 5 students in Attire
top_attire = df_cleaned.sort_values(by='Attire', ascending=False).head(5)
print('\nTop 5 students in Attire:')
display(top_attire[['Participant Name', 'Attire']])
```

### Visualizations using Plotly

```python
# Plotly visualization for Top 5 students based on Total Marks (Horizontal Bar Chart)
fig = px.bar(top_total_marks.sort_values(by='Total Marks', ascending=True),
             x='Participant Name', y='Total Marks', orientation='v',
             title='Top 5 Students by Total Marks',
             labels={'Total Marks': 'Student', 'Total Marks': 'Score'},
             color='Total Marks', color_continuous_scale=px.colors.sequential.Viridis,
             text='Total Marks') # Display values on bars
fig.update_traces(textposition='outside')
fig.update_layout(xaxis_title='Student', yaxis_title='Score')
fig.show()
```

### Average Scores Visualization

```python
# Calculate the average for each relevant column
avg_technical = df_cleaned['Technical knowledge'].mean()
avg_communication = df_cleaned['Communication'].mean()
avg_attire = df_cleaned['Attire'].mean()

# Create a DataFrame for easy plotting
avg_scores_data = pd.DataFrame({
    'Category': ['Technical Knowledge', 'Communication', 'Attire'],
    'Average Score': [avg_technical, avg_communication, avg_attire]
})

# Plot the average scores using a pie chart
fig = px.pie(avg_scores_data, values='Average Score', names='Category',
             title='Average Scores Distribution for Technical Knowledge, Communication, and Attire',
             color='Category', color_discrete_sequence=px.colors.sequential.Viridis_r)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Top 5 students in Technical knowledge

```python
# Plotly visualization for Top 5 students in Technical knowledge (Dot Plot/Scatter Plot)
fig = px.scatter(top_technical, x='Technical knowledge', y='Participant Name',
                 title='Top 5 Students by Technical Knowledge (Dot Plot)',
                 labels={'Participant Name': 'Student', 'Technical knowledge': 'Score'},
                 size='Technical knowledge', color='Technical knowledge',
                 color_continuous_scale=px.colors.sequential.Agsunset,
                 hover_name='Participant Name', text='Technical knowledge')
fig.update_traces(marker=dict(symbol='circle', size=15), textposition='middle right')
fig.update_layout(xaxis_title='Score', yaxis_title='Student',
                  xaxis_range=[min(top_technical['Technical knowledge']) - 1, max(top_technical['Technical knowledge']) + 1])
fig.show()
```

Top 5 students in Communication

```python
# Plotly visualization for Top 5 students in Communication (Pie Chart)
fig = px.pie(top_communication, values='Communication', names='Participant Name',
             title='Top 5 Students by Communication Score Distribution',
             color_discrete_sequence=px.colors.sequential.Inferno_r)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
