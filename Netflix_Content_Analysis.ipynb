Which type of content is more available on Netflix?
Which countries contribute the most content to Netflix?
Which directors have the most titles on Netflix?



```python
import pandas as pd
import numpy as np
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
content_count = df['type'].value_counts()
print(content_count)
```

```python
import matplotlib.pyplot as plt

content_count.plot(kind="bar")
plt.title("Netflix Content Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
country_count = df['country'].dropna().str.split(', ').explode().value_counts()

print(country_count.head(10))
```

```python
country_count.head(10).plot(kind='bar')
plt.title("Top 10 Countries Contributing Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
director_count = df['director'].dropna().str.split(', ').explode().value_counts()

print(director_count.head(10))
```

```python
director_count.head(10).plot(kind='bar')
plt.title("Top 10 directos with most titles on netflix")
plt.xlabel("Directors")
plt.ylabel("Title on Netflix")
plt.show()
```
