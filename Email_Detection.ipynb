```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
emails = ["Win a free lottery","Claim your cash prize","Free gift offer","Meeting at 10 Am","project submission tommorow","Complete Assessment","Lunch with friends"]
labels = ["spam","spam","spam","Not spam","Not spam","Not spam","Not spam"]
cv = CountVectorizer()
X = cv.fit_transform(emails)

model = MultinomialNB()
model.fit(X,labels)

new_email = input("Enter a new email: ")
new_email = cv.transform([new_email])
prediction = model.predict(new_email)
print("Prediction",prediction[0])
```
