Student Marks Analysis
Create a dataset containing marks of 10 students.

students = ['A','B','C','D','E','F','G','H','I','J']

marks = [78,85,92,67,88,95,72,81,90,76] Tasks

Display student marks graphically.
Find the highest scorer.
Add appropriate title and axis labels.
Display marks in descending order and plot again.

```python
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
marks = [78, 85, 92, 67, 88, 95, 72, 81, 90, 76]

plt.figure(figsize=(8, 5))
plt.bar(students, marks)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
highest_mark = max(marks)
highest_student = students[marks.index(highest_mark)]

print("Highest Scorer:", highest_student)
print("Highest Marks:", highest_mark)

sorted_data = sorted(zip(students, marks), key=lambda x: x[1], reverse=True)

sorted_students = [x[0] for x in sorted_data]
sorted_marks = [x[1] for x in sorted_data]

print("\nMarks in Descending Order:")
for s, m in sorted_data:
    print(s, ":", m)

plt.figure(figsize=(8, 5))
plt.bar(sorted_students, sorted_marks)
plt.title("Student Marks (Descending Order)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python

# Display first 5 rows
df = pd.read_csv("hospital data analysis.csv")
print(df.head())

# Check total number of patients
print("Total Patients:", len(df))

# Plot Age Distribution (Histogram)
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Create Age Groups
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Count patients in each age group
age_group_counts = df['Age_Group'].value_counts().sort_index()

print("\nAge Group Counts:")
print(age_group_counts)

# Find the most common age group
most_common = age_group_counts.idxmax()
print("\nMost Common Age Group:", most_common)

# Plot Age Groups
plt.figure(figsize=(8,5))
age_group_counts.plot(kind='bar')
plt.title("Patients by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Patients")
plt.xticks(rotation=0)
plt.show()
```
