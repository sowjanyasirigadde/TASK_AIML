Reinforcement Learning

```python
# Rewards for different actions
rewards = {
    "move_forward": 10,
    "reach_charging_station": 100,
    "hit_obstacle": -50
}

# Robot actions
actions = [
    "move_forward",
    "move_forward",
    "hit_obstacle",
    "move_forward",
    "reach_charging_station"
]

total_reward = 0

for action in actions:
    reward = rewards[action]
    total_reward += reward

    print(f"Action: {action}")
    print(f"Reward: {reward}")
    print()

print("Total Reward:", total_reward)
```

Markov Decision Process

```python
# States
states = ["Warehouse", "Road", "Customer"]

# Actions
actions = ["Move Forward", "Move Backward"]

# Rewards
rewards = {
    ("Warehouse", "Move Forward"): 10,
    ("Road", "Move Forward"): 100,
    ("Road", "Move Backward"): -10,
    ("Warehouse", "Move Backward"): -50
}

# Start state
current_state = "Warehouse"

print("Current State:", current_state)

# Robot moves forward
action = "Move Forward"
reward = rewards[(current_state, action)]

print("Action:", action)
print("Reward:", reward)

# Move to next state
current_state = "Road"
print("Next State:", current_state)
```

Q-Learning Algorithm

```python
import numpy as np

# States
states = ["Warehouse", "Middle Path", "Delivery Point"]

# Actions
actions = ["Forward", "Backward"]

# Q-Table (3 states × 2 actions)
Q = np.zeros((3, 2))

print("Initial Q-Table")
print(Q)

# Rewards
Q[0][0] = 10      # Warehouse -> Middle Path
Q[1][0] = 100     # Middle Path -> Delivery Point
Q[1][1] = -20     # Move backward
Q[2][0] = 0       # Goal reached

print("\nUpdated Q-Table")
print(Q)
```

Deep Q-Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create Neural Network
model = Sequential()

# Input layer
model.add(Dense(24, input_shape=(4,), activation='relu'))

# Hidden layer
model.add(Dense(24, activation='relu'))

# Output layer (4 possible actions)
model.add(Dense(4, activation='linear'))

# Display model
model.summary()
```
